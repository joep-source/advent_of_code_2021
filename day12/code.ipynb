```python
f = open("input.txt", "r")
input_list = f.read().split('\n')[:-1]
len(input_list)
```

```python
graph = {}
for line in input_list:
    o,d = line.split('-')
    if len(o) == 1 and len(d) == 1 and o.islower() and d.islower():
        print('skip', o, d)
        continue
    if  o!='end':
        if o not in graph.keys():
            graph[o] = []
        graph[o].append(d)
    if  d!='end':
        if d not in graph.keys():
            graph[d] = []
        graph[d].append(o)

graph
```

```python
def is_dead_end(graph, pos):
    try:
        graph[pos]
        return False
    except KeyError:
        return True

def is_repeated_move(route, pos_next):
    try:
        if pos_next.islower() and pos_next in route:
            return True
        return False
    except IndexError:
        return False

def is_new_route(routes, route, pos_next):
    route_new = route + [pos_next]
    for route_exist in reversed(routes):
        if route_new == route_exist[:len(route_new)]:
            return False
    return True

def complete_last_route(graph, routes):
    route = routes[-1]
    for _ in range(10000):
        if route[-1] == 'end':
            return routes
        pos = route[-1]                
        for pos_next in graph[pos]:
            if (not is_repeated_move(route, pos_next)) and is_new_route(routes, route, pos_next):
                route.append(pos_next)
                break
    return routes
         

def add_new_route(graph, routes):
    route = routes[-1]
    for i in reversed(range(len(route)-1)):
        pos = route[i]
        new_route=route[:i+1]
        if pos in graph.keys():            
            for pos_option in graph[pos][1:]:
                if (not is_repeated_move(new_route, pos_option)) and is_new_route(routes, new_route, pos_option):
                    new_route.append(pos_option)
                    routes.append(new_route)
                    return routes        
    print(sum([r[-1]=="end" for r in routes]))
    print('THE END')
    raise EOFError

                
routes = [['start']]
for _ in range(10000):
    routes = complete_last_route(graph, routes)    
    print(len(routes), sum([r[-1]=="end" for r in routes]))
    routes = add_new_route(graph, routes) 
```
